# Problema da fabrica de papeis

## Descrição do problema

- Uma pequena fábrica de papel toalha manufatura três tipos de produtos A, B e C. 
- A fábrica recebe o papel em grandes rolos.
- O papel é cortado, dobrado e empacotado. 
- O lucro unitário de cada produto é respectivamente R$ 10,00, R$ 15,00 e R$ 20,00. 
- O quadro abaixo identifica o tempo requerido (em horas) em cada seção da fábrica para produção de uma unidade de cada item, bem como a quantidade de máquinas disponı́veis, que trabalham 40 horas por semana.
- Desejamos planejar a produção semanal da fábrica com objetivo de temos o lucro máximo, usando programação linear.


| Seção         | Produto A | Produto B | Produto C | Qtd de Máquinas |
|---------------|-----------|-----------|-----------|-----------------|
| Corte         |         8 |         5 |         2 |               3 |
| Dobra         |         5 |        10 |         4 |              10 |
| Empacotamento |       0,8 |         1 |         2 |               2 |

### Variáveis

São o número de unidades a serem fabricadas, semanalmente, dos produtos A, B e C. 

Sejam tais variáveis $x_A \geq 0$ , $x_B \geq 0$ e $x_C \geq 0$, respetivamente.

### Função objetivo

A função objetivo expressa o lucro a ser maximizado, sendo portanto
$
f(x_A , x_B , x_C ) = 10 x_A + 15 x_B + 20 x_C
$

### Restrições

Temos uma restrição para cada seção da fábrica, estabelecendo que o limite de horas semanais disponı́veis na seção não seja excedido com a produção.

#### Seção corte
- Produzindo $x_A$ unidades de A, usamos $8 x_A$ horas 
- Produzindo $x_B$ unidades de B, usamos $5 x_B$ horas
- Produzindo $x_C$ unidades de B, usamos $2 x_C$ horas
- Considerando que há 3 máquinas para corte, totalizando 3 x 40 horas disponı́veis nessa seção, assim temos
$
8x_A + 5x_B + 2x_C \leq 120
$

### Seção de dobra 
- Produzindo $x_A$ unidades de A, usamos $5 x_A$ horas 
- Produzindo $x_B$ unidades de B, usamos $10 x_B$ horas
- Produzindo $x_C$ unidades de B, usamos $4 x_C$ horas
- Há 10 máquinas para dobra, totalizando 10 x 40 horas disponı́veis nessa seção, assim temos
$
5 x_A + 10 x_B + 4 x_C \leq 400
$

### Seção de empacotamento
- Produzindo $x_A$ unidades de A, usamos $0,8  x_A$ horas 
- Produzindo $x_B$ unidades de B, usamos $1 x_B$ horas
- Produzindo $x_C$ unidades de B, usamos $2 x_C$ horas
- Há 10 máquinas para , totalizando 2 x 40 horas disponı́veis nessa seção, assim temos
$
0,8 x_A + 1 x_B + 2 x_C \leq 80
$

### Observações

Não definimos no enunciado a unidade em que é medida a produção dos itens A, B e C. 

Por exemplo, caso seja quilograma (Kg), podemos considerar que as variáveis $x_A$ , $x_B$ e $x_C$ admitem valores reais. 

Nesse caso, temos, de fato, um problema de programação linear.

Entretanto, caso a unidade fosse número de caixas, poderı́amos considerar que as variáveis só admitiriam valores inteiros. 

Para isso, terı́amos que incluir as restrições de integralidade $x_A, x_B, x_C \in \mathbb{Z}$. 

Do ponto de vista de modelagem, essa é uma pequena modificação. Porém, a inclusão dessas restrições torna a resolução do modelo bem mais complexa.

```python
# importando bibliotecas
import gurobipy as gp
from gurobipy import GRB
```

```python
# cria model
model = gp.Model("fabrica_papeis_lp")
```

```python
model
```

```python
# define variaveis
xA = model.addVar(name="xA")
xB = model.addVar(name="xB")
xC = model.addVar(name="xC")
model.update()
```

```python
model
```

```python
# define funcao objetivo
model.setObjective(10*xA + 15*xB + 20*xC, GRB.MAXIMIZE)
model.update()
```

```python
# define restricoes
con1 = model.addConstr(8*xA + 5*xB + 2*xC  <= 120, "cortes")
con2 = model.addConstr(5*xA + 10*xB + 4*xC  <= 400, "dobras")
con3 = model.addConstr(0.8*xA + 1*xB + 2*xC <= 80, "empacotamento")
model.update()
```

```python
model
```

```python
# export lp/mps
model.write('fabrica_papeis_lp.mps')
```

```python
# otimiza o modelo
model.optimize()
```

```python
# imprime solucao ótima
for v in model.getVars():
    print(f"{v.VarName} = {v.X}")
```

```python
# imprime valor ótimo
print(f"obj = {model.ObjVal}")
```

```python
# imprime tempo de execucao
print(f'Runtime: {model.Runtime}')
```

```python
# limpa todos os recursos associados ao objeto model
model.dispose()
```

## Versão programação inteira

```python
# cria model
model = gp.Model("fabrica_papeis_lp")

# variaveis
xA = model.addVar(vtype=GRB.INTEGER, name="xA")
xB = model.addVar(vtype=GRB.INTEGER, name="xB")
xC = model.addVar(vtype=GRB.INTEGER, name="xC")

# funcao objetivo
model.setObjective(10*xA + 15*xB + 20*xC, GRB.MAXIMIZE)

# restricoes
con1 = model.addConstr(8*xA + 5*xB + 2*xC  <= 120, "cortes")
con2 = model.addConstr(5*xA + 10*xB + 4*xC  <= 400, "dobras")
con3 = model.addConstr(0.8*xA + 1*xB + 2*xC <= 80, "empacotamento")

# export lp/mps
model.write('fabrica_papeis_ip.mps')

# otimiza o modelo
model.optimize()
```

```python
# imprime solucao ótima
for v in model.getVars():
    print(f"{v.VarName} = {v.X}")
```

```python
# imprime tempo de execucao
print(f'Runtime: {model.Runtime}')
```

```python
# limpa todos os recursos associados ao objeto model
model.dispose()
```
